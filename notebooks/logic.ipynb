# Propositional Logic Test
Similar to Graves, Alex. "Adaptive Computation Time for Recurrent Neural Networks." arXiv preprint arXiv:1603.08983 (2016).
## Data Generation

```python
import pandas as pd

gates = [
    0, 
    1,
    lambda x,y: not (x or y),
    lambda x,y: not x and y,
    lambda x,y: x and not y,
    lambda x,y: (x or y) and not(x and y),
    lambda x,y: not (x and y),
    lambda x,y: x and y,
    lambda x,y: not((x or y) and not(x and y)),        
    lambda x,y: not x or y,
    lambda x,y: x or not y,
    lambda x,y: x or y
]
vocab = {
    True: 0,
    False: 1,
    "NOR": 2,
    "Xq": 3,
    "ABJ": 4,
    "XOR": 5,
    "NAND": 6,
    "AND": 7,
    "XNOR": 8,
    "if/then": 9,
    "then/if": 10,
    "OR": 11
}

columns = ["P", "Q", "NOR", "Xq", "ABJ", "XOR", "NAND", "AND", "XNOR", "if/then", "then/if", "OR"]
index = ["TT", "TF", "FT", "FF"]
df = pd.DataFrame(columns=columns, index=index)
df["P"]["TT"] = "T"
df["Q"]["TT"] = "T"
df["P"]["TF"] = "T"
df["Q"]["TF"] = "F"
df["P"]["FT"] = "F"
df["Q"]["FT"] = "T"
df["P"]["FF"] = "F"
df["Q"]["FF"] = "F"

for gate in columns[2:]:
    for input in index:
        x = df["P"][input] == "T"
        y = df["Q"][input] == "T"  
        df[gate][input] = str(gates[vocab[gate]](x,y))[0]
df
```
